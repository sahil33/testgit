# Import necessary Packages

```
import numpy as np
import cv2
import face_recognition
from PIL import Image
import matplotlib.pyplot as plt
```

# Import an image for face recognition and show that image

```
image = face_recognition.load_image_file("President_Barack_Obama.jpg")
plt.imshow(image)
plt.show();
```

# Run face_regconition package to identify face locations coordinate in the Image using convolution Neural Network(CNN)

```
face_locations = face_recognition.face_locations(image,model="cnn")
face_locations
```

# Draw a blue box around the face

```
for face_location in face_locations:
  top,right,bottom,left = face_location
  img = cv2.rectangle(image,(left,top),(right,bottom),(0,0,256),2)
  
plt.imshow(img)
plt.show();
```

# Using face coordinates cut out the face only

```
for face_location in face_locations:
  
  top,right,bottom,left = face_location
  face_image = image[top:bottom, left:right]
  pil_image = Image.fromarray(face_image)
  
plt.imshow(pil_image)
plt.show();
```
